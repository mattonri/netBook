```python
##START UP
import sqlite3
import pandas as pd
import numpy as np

# Connect to the SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('netbookdb.db', timeout=10)

# Create a cursor object to execute SQL queries
cursor = conn.cursor()
```

```python
#import data from xsl
df = pd.read_excel('careerCenterInternships.xlsx')

# df.info()
df.head()
df.columns
```

```python
#Drop department and college columns
df = df.drop(['College', 'Department'], axis=1)

#Rename Columns
df.rename(columns={'Supervisor Name': 'Supervisor_Name'}, inplace=True)
df.rename(columns={'Supervisor Title': 'Supervisor_Title'}, inplace=True)
df.rename(columns={'Supervisor Email': 'Supervisor_Email'}, inplace=True)
df.rename(columns={'Course Number': 'Course_Number'}, inplace=True)
df.rename(columns={'Employer City': 'Employer_City'}, inplace=True)
df.rename(columns={'Employer State': 'Employer_State'}, inplace=True)
df.rename(columns={'Employer Country': 'Employer_Country'}, inplace=True)
df.rename(columns={'Job City': 'Job_City'}, inplace=True)
df.rename(columns={'Job State': 'Job_State'}, inplace=True)
df.rename(columns={'Job Country': 'Job_Country'}, inplace=True)
df.rename(columns={'Faculty Email': 'Faculty_Email'}, inplace=True)
df.dropna(subset=['Supervisor_Email'], inplace=True)
location_columns = ["Employer_City", "Employer_State", 'Employer_Country', 'Job_City', 'Job_State', 'Job_Country', 'Faculty_Email', 'Supervisor_Title', 'Job Title', 'Employer Name', 'Course_Number']
for column in location_columns:
    df[column].fillna('NA', inplace=True)

df
```

```python
#INSERT INTO countries
job_countries = df['Employer Country'].unique().tolist()
employer_countries = df['Job Country'].unique().tolist()
for c in job_countries:
    if c not in employer_countries:
        employer_countries.append(c)

employer_countries.pop(1)
print(employer_countries)

for name in employer_countries:
    sql = 'INSERT INTO countries (name) VALUES (?)'
    cursor.execute(sql, (name,))  # Use a tuple with a comma to pass the parameter
    conn.commit()
```

```python
#INSERT INTO locations
locations = set()
for index, row in df.iterrows():
    city = str(row['Employer_City']).title()
    state = str(row['Employer_State']).title()
    country = str(row['Employer_Country'])
    city2 = str(row['Job_City']).title()
    state2 = str(row['Job_State']).title()
    country2 = str(row['Job_Country'])
    l = (city, state, country)
    l2 = (city2, state2, country2)
    for i in range(len(l)):
        if l[i] == 'Na':
            l[i].upper()
    for i in range(len(l2)):
        if l2[i] == 'Na':
            l2[i].upper()
    locations.add(l)
    locations.add(l2)

print(len(locations))
counter = 0
for l in locations:
    counter += 1
    country_name = l[2]
    cursor.execute(f"SELECT country_id FROM countries WHERE countries.name == ?", (country_name,))
    print(country_name)
    result = cursor.fetchone()[0]
    cursor.execute(f'INSERT INTO locations (city, state, country_id) VALUES (?, ?, ?)', (l[0], l[1], result))
    print(f"Completed {counter}")
    conn.commit()


conn.commit()
```

```python
#INSERT INTO companies
companies = set()
for index, row in df.iterrows():
    city = str(row['Employer_City']).title()
    state = str(row['Employer_State']).title()
    country = str(row['Employer_Country'])
    company_name = str(row['Employer Name'])
    companies.add((city, state, country, company_name))


for count, company in enumerate(companies):
    cursor.execute(f"SELECT locations.location_id FROM locations JOIN countries ON locations.country_id = countries.country_id WHERE locations.city = ? AND locations.state = ? AND countries.name = ?", (company[0], company[1], company[2]))
    result = cursor.fetchone()[0]
    cursor.execute(f'INSERT INTO companies (name, location_id) VALUES (?, ?)', (company[3], result))
    if count % 1000 == 0:
        conn.commit()
        print(f'Completed {count}')
```

```python
#INSERT INTO supervisors
supervisors = set()
for index, row in df.iterrows():
    name = row['Supervisor_Name']
    title = row['Supervisor_Title']
    email = row['Supervisor_Email']
    company_name = row['Employer Name']
    city = str(row['Employer_City']).title()
    state = str(row['Employer_State']).title()
    country = str(row['Employer_Country'])
    tem = [city, state, country]
    for i in range(len(tem)):
        if tem[i] == 'Na':
            tem[i] = tem[i].upper()
    bundle = (name, title, email, company_name, tem[0], tem[1], tem[2])
    supervisors.add(bundle)


for count, s in enumerate(supervisors):
    #query for company_id
    sql = """SELECT companies.company_id FROM companies 
    JOIN locations ON companies.location_id = locations.location_id 
    JOIN countries ON locations.country_id = countries.country_id 
    WHERE companies.name = ? AND locations.city = ? AND locations.state = ? AND countries.name = ?
    """
    cursor.execute(sql, (s[3], s[4], s[5], s[6]))
    result = cursor.fetchone()[0]
    cursor.execute(f'INSERT INTO supervisors (name, title, email, company_id) VALUES (?, ?, ?, ?)', (s[0], s[1], s[2], result))
    if count % 1000 == 0:
        conn.commit()
        print(f'Completed {count}')
```

```python
#INSERT INTO internships
for index, row in df.iterrows():
    #find info for major_id
    major = row[0]
    #find info for course_id
    course = row[1]
    #find info for faculty_id
    fac_name = row[2]
    fac_email = row[3]
    credits = row[4]
    semester = row[5]
    hours = row[6]
    compensations = row[7]
    start_date = str(row[8])
    end_date = str(row[9])
    title = row[17]
    #find info for location_id (the job location)
    job_city = row[14].title()
    job_state = row[15].title()
    job_country = row[16]
    #find info for supervisor_id
    sup_name = row[18]
    sup_title = row[19]
    sup_email = row[20]
    #find info for company_id to find super_id
    comp_name = row[10]
    #find info for location_id to company_id
    comp_city = row[11].title()
    comp_state = row[12].title()
    comp_country = row[13]

    temp = [job_city, job_state, comp_city, comp_state]
    for item in range(len(temp)):
        if temp[item] == 'Na':
            temp[item] = temp[item].upper()

    job_city = temp[0]
    job_state = temp[1]
    comp_city = temp[2]
    comp_state = temp[3]
    
    # print(course)

    major_id_sql = """SELECT major_id FROM majors WHERE name = ?"""
    cursor.execute(major_id_sql, (major,))
    major_id = cursor.fetchone()[0]

    course_id_sql = """SELECT course_id FROM courses WHERE course_code = ?"""
    cursor.execute(course_id_sql, (course,))
    course_id = cursor.fetchone()[0]

    faculty_id_sql = """SELECT faculty_id FROM faculty WHERE name = ? AND email = ?"""
    cursor.execute(faculty_id_sql, (fac_name, fac_email))
    faculty_id = cursor.fetchone()[0]

    supervisor_id_sql = """SELECT supervisor_id FROM supervisors 
    JOIN companies ON supervisors.company_id = companies.company_id
    JOIN locations ON companies.location_id = locations.location_id
    JOIN countries ON countries.country_id = locations.country_id
    WHERE supervisors.name = ? AND supervisors.title = ? AND supervisors.email = ? AND companies.name = ? AND locations.city = ? AND locations.state = ? AND countries.name = ?
    """
    cursor.execute(supervisor_id_sql, (sup_name, sup_title, sup_email, comp_name, comp_city, comp_state, comp_country))
    supervisor_id = cursor.fetchone()[0]

    location_id_sql = """SELECT location_id FROM locations
    JOIN countries ON locations.country_id = countries.country_id
    WHERE city = ? AND state = ? AND countries.name = ?"""
    cursor.execute(location_id_sql, (job_city, job_state, job_country))
    location_id = cursor.fetchone()[0]

    internship_sql = """INSERT INTO internships (major_id, course_id, faculty_id, credits, semester, weekly_hours, compensation, start_date, end_date, title, supervisor_id, location_id) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)"""
    cursor.execute(internship_sql, (major_id, course_id, faculty_id, credits, semester, hours, compensations, start_date, end_date, title, supervisor_id, location_id))
    
    if index % 1000 == 0:
        conn.commit()
        print(f'Completed {index}')
    
```
